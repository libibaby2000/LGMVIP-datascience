**Image to Pencil Sketch with Python**

**Description Of Task :**We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image. Since images are just arrays, we can easily do this programmatically using the divide function from the cv2 library in Python.


```python
import cv2
from matplotlib import pyplot as plt
```

**Read the image**

```python
ims = cv2.imread("Rabbit.jpg")
plt.figure(figsize=(8,8))
plt.imshow(ims)
plt.show()
```

**Converting the image to rgb colors**

```python
rgimg = cv2.cvtColor(ims, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(rgimg)
plt.show()
```

**Converting the rgb image in to a gray scale image**

```python
gray_img = cv2.cvtColor(ims, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_img,cmap='gray')
plt.show()
```

**Converting the gray scale image in to a inverted image**

```python
invert_gray_img =255 - gray_img
plt.figure(figsize=(8,8))
plt.imshow(invert_gray_img,cmap='gray')
plt.show()
```

**Blur the inverted grayscale image**

```python
blur_img = cv2.GaussianBlur(invert_gray_img, (21,21), 0)
plt.figure(figsize=(8,8))
plt.imshow(blur_img,cmap='gray')
plt.show()
```

**Invert the blurred image**



```python
invert_blur_img = 255 - blur_img
plt.figure(figsize=(8,8))
plt.imshow(invert_blur_img,cmap='gray')
plt.show()
```

**Divide the grayscale image with inverted blurred image to get the sketch**

```python
sketch = cv2.divide(gray_img, invert_blur_img, scale=256.0)
plt.figure(figsize=(8,8))
plt.imshow(ims)
plt.figure(figsize=(8,8))
plt.imshow(sketch,cmap='gray')
plt.show()
```
